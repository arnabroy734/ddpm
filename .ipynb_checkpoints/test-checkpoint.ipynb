### Test Noise Scheduler

```python
path = Path.cwd()/'data/celeb-dataset/img_align_celeba/img_align_celeba'
for file in path.iterdir():
    image = Image.open(file)
    break
# image
```

```python
loader = get_dataloader(datapath='data/celeb-dataset/img_align_celeba/img_align_celeba', batch= 10)
```

```python
X0 = None
for X in loader:
    X0 = X
    X0 = X0.to('cuda')
    break
```

```python
t = torch.randint(low=0, high=1000, size=(10, )).to('cuda')
t
```

```python
noise = torch.randn(X0.shape).to('cuda')
```

```python
scheduler = LinearNoiseScheduler()
```

```python
Xt = scheduler.add_noise(X0, t, noise)
# Xt = (Xt + 1)/2
Xt1 = torch.clamp((Xt+1)/2, 0, 1)
```

```python
transforms.ToPILImage()(Xt1[5])
```

```python
x_t_1, x0 = scheduler.sample(Xt, 748, noise)
```

```python
transforms.ToPILImage()(torch.clamp((x0+1)/2, 0, 1)[1])
```
